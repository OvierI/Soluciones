```python
x = sympy.Symbol('x')
y = sympy.Function('y')
```

```python
diff(y(x))-5*y(x)-x/3
```

```python
Eq(diff(y(x))-5*y(x),x/3)
```

```python
ED=Eq(diff(y(x))-5*y(x),x/3)
```

```python
dsolve(ED,y(x))
```

```python
ED2=Eq(diff(y(x))-5*y(x),8)
```

```python
CI={y(0):10}
```

```python
dsolve(ED2,y(x),ics=CI)
```

```python
diff(y(x),x,2)+2*x**3
```

```python
Eq(diff(y(x),x,2),-2*x**3)
```

```python
ED3=Eq(diff(y(x),x,2),-2*x**3)
```

```python
CI={y(0):0,y(10):0}
```

```python
dsolve(ED3,y(x),ics=CI)
```
